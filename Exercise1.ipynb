# Quantification of COVID-19 Vaccine Coercion in India

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('table1.csv')

df_filtered = df[df['Survey response'] != 'Total']

categories = df_filtered['Survey response']
frequency = df_filtered['Frequency']
percentage = df_filtered['Percentage']

plt.figure(figsize=(8, 4))

bars = plt.bar(categories, frequency, color=['red', 'green', 'blue'])

for bar, freq in zip(bars, frequency):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, 
             bar.get_height() + 0.2, str(freq), 
             ha='center', color='black')

total_count = df.loc[df['Survey response'] == 'Total', 'Frequency'].values[0]
total_percentage = df.loc[df['Survey response'] == 'Total', 'Percentage'].values[0]

plt.text(2, max(frequency) + 6, f'Total: {total_count}', ha='left', color='black', fontsize=12)


plt.title('Survey Responses from Non-Student Adults')
plt.xlabel('Survey Response')
plt.ylabel('Frequency')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('table3.csv')

colors = ['#ff9999', '#66b3ff', '#99ff99']

fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(df['Frequency'], labels=df['Survey response'], autopct=lambda p: f'{p:.2f}%  ({int(round(p * sum(df["Frequency"]) / 100))})',
                                  startangle=90, colors=colors, wedgeprops=dict(width=0.3, edgecolor='w'))

ax.axis('equal') 

plt.title('Survey Responses')

for text, autotext in zip(texts, autotexts):
    text.set_size(8)
    autotext.set_size(6)

# Display the chart
plt.show()
```
